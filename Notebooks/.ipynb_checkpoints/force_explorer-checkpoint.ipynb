```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python

distance = np.arange(1.1, 2, 0.001)
Force = []
epsilon = 5
r_vector = np.zeros([1, 2])
r_vector[0, 0] = 1
for i in distance:
    LJ = 24*epsilon*(-28*(i**-13) +8*(i**-7) )
#     print(LJ)
    Force.append(LJ)

plt.plot(distance, Force)
plt.xlabel("distance")
plt.ylabel("force")
plt.show()

```
